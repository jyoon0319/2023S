## 워드클라우드를 위한 전처리
1. title과 description 합치기
2. 불필요한 기호/태그 등 삭제

```python
# 'items'에 들어있는 title과 description 데이터 추출하기
for item in jsonObj['items'] :
  print(item['title'])
  print(item['description'])
```

```python
# 합쳐서 리스트에 담기
result = []

for item in jsonObj['items'] :
  result.append(item['title'])
  result.append(item['description'])

print(len(result))
```

```python
# 불필요한 문자/기호 등 삭제
result = []

result = []
for item in jsonObj['items'] :
  str_temp = item['title'].replace('""', '').replace('#', '').replace('...','').replace('?', '')
  str_temp = item['title'].replace('!', '').replace('<b>', '').replace('</b>','').replace('[', '').replace(']','')
  result.append(str_temp)

for item in jsonObj['items'] :
  str_temp = item['description'].replace('""', '').replace('#', '').replace('...','').replace('?', '')
  str_temp = item['description'].replace('!', '').replace('<b>', '').replace('</b>','').replace('[', '').replace(']','')
  result.append(str_temp)
print(len(result))
print(result)
```

```python
#워드클라우드를 위해 문자열 모두 합치기
result_str = ''

for r in result :
  result_str = result_str + r    #result_str += r
print(result_str)
```

```python
word_dict = {}

for word, w_count in n_count.most_common(100):   #상위 빈도 100개의 단어 & 글자의 길이가 2 이상인 단어들만 추출해서 딕셔너리에 저장하기
    if len(word) >= 2 : 
        word_dict[word] = w_count
print(word_dict )
```
